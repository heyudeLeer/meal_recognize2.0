## pixclass description

### "pixclass" is a python library of meican.com for pixel classification, It is implemented by convolutional neural network , provides two interfaces for webapp engineers or product engineers for image recognize.

```python
from ImgSeg import pixclass

restaurant_path = '/home/heyude/PycharmProjects/mysite/meal/restaurant/kfc'

#train model first of all
model_info = pixclass.train_data_set(data_set_path=restaurant_path)

# hot predict by model_info, for saving time
ret = pixclass.hot_predict_img(data_set_path=restaurant_path,
                               image_url=restaurant_path + '/predictImg/qs-fqjd1.jpg',
                               predict_info=model_info)

# init start or change the restaurant
ret, model_info = pixclass.cold_predict_img(data_set_path=restaurant_path,
                                            image_url=restaurant_path + '/predictImg/qs-fqjd2.jpg')
```
